```python
# Pernyataan 1: 1 + 3 + 5 + ... + (2n-1) = n²
print("Pernyataan 1: 1 + 3 + 5 + ... + (2n-1) = n²")
print("Rentang uji: n = 0 sampai 20")
valid1 = []
for n in range(0, 21):
    left = sum(2*i-1 for i in range(1, n+1)) if n > 0 else 0
    right = n**2
    result = left == right
    if result: valid1.append(n)
    print(f"n={n}: {left} = {right} ({result})")

print(f"n valid: {valid1}")

# Pernyataan 5: 2ⁿ > n²
print("\nPernyataan 5: 2ⁿ > n²")
print("Rentang uji: n = -10 sampai 10")
valid5 = []
for n in range(-10, 11):
    left = 2**n
    right = n**2
    result = left > right
    if result: valid5.append(n)
    print(f"n={n}: {left:.3f} > {right} ({result})")

print(f"n valid: {valid5}")

# Kesimpulan
print("\n" + "="*40)
print("KESIMPULAN:")
print(f"Pernyataan 1 valid untuk n: {valid1}")
print(f"Pernyataan 5 valid untuk n: {valid5}")
print("\nKedua pernyataan memiliki jangka valid yang berbeda")
print("Pernyataan 1: Dengan menggunakan syarat n >= 0 dan dilakukan uji sampai n=20, terlihat bahwa semua nilainya benar. Jadi semua n >= 0 adalah benar")
print("Pernyataan 5: Dengan menggunakan syarat n>=-10 dan dilakukan uji sampai n=10, terlihat bahwa yang benar hanya jika n= (0, 1, 5, 6, 7, 8, 9, 10). Jadi pernyataan 5 benar untuk semua n >= 5 dan n=0 dan n=1 ")
```
