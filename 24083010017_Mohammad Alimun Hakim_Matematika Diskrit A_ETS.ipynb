```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

print("="*90)
print(" "*20 + "TUGAS PENGGANTI UTS - MATEMATIKA DISKRIT (A)")
print("="*90)
print()

# ===================================
# BAGIAN 1: INFORMASI ARTIKEL ILMIAH 
# ===================================
print("╔" + "="*88 + "╗")
print("║" + " "*25 + "1. INFORMASI ARTIKEL ILMIAH" + " "*36 + "║")
print("╚" + "="*88 + "╝")
print()

artikel_info = {
    "Judul": "Penerapan Data Mining Untuk Memprediksi Ketersediaan Stok\n Produk HNI HPAI Menggunakan Algoritma C4.5",
    "Penulis": "Murlena, Diwi Apriana",
    "Institusi": "Universitas Pat Petulai, Bengkulu, Indonesia",
    "Corresponding Author": "Murlena",
    "Email": "ana.murlena@gmail.com",
    "Negara": "INDONESIA",
    "DOI": " https://doi.org/10.29240/arcitech.v2i1.5271",
    "Status DOI": "VALID (Dapat diakses)",
    "Jurnal": "Arcitech: Journal of Computer Science and Artificial Intelligence",
    "Volume": "Volume 2 No. 1 (January - June 2022)",
    "Halaman": "pp. 19-32",
    "Bahasa": "Indonesia",
    "Tanggal Submit": "8 April 2022",
    "Tanggal Publikasi": "25 Juni 2022"
}

print("DETAIL ARTIKEL:")
print("-" * 90)
for k, v in artikel_info.items():
    if k == "Judul":
        print(f"{k:20}: {v}")
    else:
        print(f"{k:20}: {v}")

print("\n" + "─" * 10)
print("ABSTRAK")
print("─" * 10)

# ==============
# TEKS LENGKAP 
# ==============
p1_abstrak = """Pada suatu perusahaan, ketersediaan stok ini harus diperhatikan untuk 
menghindari kerugian produk yang disebabkan karena expired dan juga untuk 
menghindari terjadinya pembatalan orderan. Data yang terdapat pada aplikasi 
perusahaan HPAI yang dinamakan aplikasi HSIS ini dapat dimanfaatkan dan 
diterapkannya data mining dalam menentukan produk-produk apa saja yang laris 
dalam kehidupan sehari-hari. Untuk menjaga ketersediaan stok produk, maka 
tujuan penelitian ini adalah melakukan pengujian pada AgenStok menggunakan 
Algoritma C4.5 untuk memprediksi agar produk yang harus dijaga 
ketersediaanya sesuai dengan kebutuhan customer. Metode penelitian yang 
gunakan dalam penelitian ini menggunakan teknik Data Mining dengan analisis 
menggunakan algoritma C4.5. Diharapkan penggunaan Analisis menggunakan 
Algoritma C4.5 pada Agenstok bisa mengatasi penumpukan stok dan keuangan 
atau modal Agenstok dapat dimanfaatkan untuk membeli stok yang harus 
tersedia. Hasil dari analisa menggunakan algoritma C4.5 menyatakan bahwa 
meskipun harga produknya mahal namun tetap laris adalah herbs product (produk 
herbal). Jadi dari ketiga jenis produk HNI HPAI (Herbs product, Health food & 
beverage dan Cosmetic & home care) maka yang stoknya harus dibanyakkan atau 
tersedia adalah Herbs Product."""

print(p1_abstrak)

print("\n" + "─" * 28)
print("PENDAHULUAN (Introduction)")
print("─" * 28)

p2_pendahuluan_1 = """Perkembangan teknologi dan komunikasi saat ini menuntut semua aspek untuk 
beradaptasi dalam semua bidang seperti bidang pendidikan, industri, ekonomi serta 
perdagangan (Widyastuti et al. 2021). Produk HPAI ini merupakan produk herbal yang 
saat ini laris dipasaran karena produk-produknya diakui bagus dan berkwalitas. Untuk 
lebih meningkatnya kinerja pada perusahaan dibidang pelayanan dan kemudahan dalam 
mengakses informasi dan juga transaksi yang real time bagi agen dan stakeholder maka 
HNI-HPAI menggunakan beberapa teknologi, salah satunya HSIS (HNI-HPAI Sales 
Integrated System). HSIS ini mengintegrasikan transaksi yang dilakukan secara online 
dengan beberapa fitur dan informasi yang mudah diakses secara real time diantaranya 
mengenai peningkatan omzet, berapa banyak ketersediaan saldo produk dan juga 
bagaimana perkembangan jumlah agen per hari."""

p3_pendahuluan_2 = """Dikarenakan produk-produk herbal ini ada batas penggunaannya maka sebelum 
expired produk tersebut harus habis. Untuk mengatasi agar tidak terjadinya penumpukan 
produk pada agenstok maka dilakukan penerapan data mining untuk memprediksi 
ketersediaan stok. Dengan tidak menumpuknya produk pada Agenstok maka 
perputaran keuangan atau modal pun dapat teratasi (Fikri and Verina 2020). Maka 
dengan menggunakan Algoritma C4.5 ini dapat diprediksi produk apa saja yang harus 
disediakan lebih banyak dan produk apa yang tidak mesti disediakan, sehingga dengan 
stok yang sesuai dapat memenuhi kebutuhan konsumen dan agenstok pun terhindar dari 
pembatalan orderan yang dapat merugikan agenstok itu sendiri (Jumiyati and 
Prasetyaningrum 2020)."""

print(p2_pendahuluan_1)
print()
print(p3_pendahuluan_2)

print("\n" + "─" * 25)
print("KESIMPULAN (Conclusion)")
print("─" * 25)

p4_kesimpulan_1 = """Berdasarkan uraian yang telah dilakukan pada analisa diatas, dapat diambil 
kesimpulan bahwa berdasarkan analisa yang dilakukan terhadap produk HNI HPAI 
yang jumlah terjualnya banyak meskipun harganya tergolong mahal adalah produk 
herbal (herbs product). Sesuai dengan pohon keputusan hasil perhitungan node 1.1.2 maka 
produk herbal (herbs product) yang harus di stok lebih banyak dibandingkan dengan 
health food and beverage. Sehingga dengan penerapan data mining menggunakan 
algoritma C4.5 ini maka para agenstok dapat terhindar dari penumpukan stok yang 
kurang laris."""

p5_kesimpulan_2 = """Data-data penjualan dan pembelian pada Business Center (BC) tentu lebih 
banyak lagi dibandingkan data pada Agenstok, maka penerapan algoritma C4.5 ini juga 
bisa dikembangkan pada Business Center (BC) dalam menentukan stok barang 
khususnya pada produk herbalnya yang laris berdasarkan mereknya (nama-nama produk 
herbal) agar sesuai dengan kebutuhan Agenstok atau konsumen."""

print(p4_kesimpulan_1)
print()
print(p5_kesimpulan_2)

# Gabung semua paragraf untuk analisis berikutnya
all_paragraphs = [p1_abstrak, p2_pendahuluan_1, p3_pendahuluan_2, p4_kesimpulan_1, p5_kesimpulan_2]
full_text = " ".join(all_paragraphs)

print()

# ===============
# CLEANING TEKS
# ===============
def extract_words(text):
    # ubah tanda hubung, garis bawah, dan slash menjadi spasi agar kata terpisah
    cleaned = re.sub(r'[-_/]', ' ', text)
    # hapus kata yang mengandung huruf dan angka sekaligus (misal C4.5, H2O)
    cleaned = re.sub(r'\b[a-zA-Z]*\d+[a-zA-Z\d.]*\b', ' ', cleaned)
    # hapus semua karakter non-huruf (selain spasi)
    cleaned = re.sub(r'[^a-zA-Z\s]', ' ', cleaned)
    # ambil hanya kata huruf (min 2 huruf)
    words = re.findall(r'\b[a-zA-Z]{2,}\b', cleaned)
    # kembalikan dalam huruf kecil semua
    return [w.lower() for w in words]
    
def count_cv(word):
    """Hitung jumlah konsonan dan vokal dalam kata."""
    vowels = "aiueo"
    consonants = sum(1 for c in word if c not in vowels and c.isalpha())
    vowel_count = sum(1 for c in word if c in vowels)
    return consonants, vowel_count
    
def extract_sentences(text):
    sentences = re.split(r'(?<=[.!?])\s+(?=[A-Z])', text)
    return [s.strip() for s in sentences if len(s.strip()) > 10]
    
# ============================================================================
# BAGIAN 2: TABEL PIVOT KONSONAN DAN VOKAL 
# ============================================================================
print("╔" + "="*88 + "╗")
print("║" + " "*20 + "2. ANALISIS KONSONAN DAN VOKAL DALAM KATA" + " "*27 + "║")
print("╚" + "="*88 + "╝")
print()

# Ekstraksi kata
words = extract_words(full_text)

# === DF utama: kata + struktur CV (konsonan, vokal)
df_words = pd.DataFrame({'word': words})
df_words[['konsonan', 'vokal']] = df_words['word'].map(count_cv).apply(pd.Series)

# Pivot 1 — FREKUENSI total (semua kemunculan)
pivot_frequency = pd.crosstab(df_words['vokal'], df_words['konsonan'])

# Pivot 2 — CACAH KATA UNIK (tiap kata dihitung 1×)
df_unique = df_words.drop_duplicates(subset='word')
pivot_count = pd.crosstab(df_unique['vokal'], df_unique['konsonan'])

print(" Tabel 1: FREKUENSI Total Kombinasi Konsonan–Vokal")
print("-" * 90)
print(pivot_frequency)
print()

print("Tabel 2: CACAH KATA Unik per Kombinasi Konsonan–Vokal")
print("-" * 90)
print(pivot_count)
print()

# Statistik ringkas
print("STATISTIK DESKRIPTIF:")
print("-" * 90)
print(f"Total kata dianalisis       : {len(df_words):,}")
print(f"Total kata unik             : {len(df_unique):,}")
print(f"Rata-rata konsonan/kata     : {df_words['konsonan'].mean():.2f}")
print(f"Rata-rata vokal/kata        : {df_words['vokal'].mean():.2f}")
ratio_cv = df_words['konsonan'].mean() / (df_words['vokal'].mean() if df_words['vokal'].mean() != 0 else 1)
print(f"Rasio konsonan:vokal        : {ratio_cv:.2f}:1")

mode_combo = df_words.groupby(['konsonan', 'vokal']).size().idxmax()
top_combinations = (df_words.groupby(['konsonan','vokal'])
                    .size().reset_index(name='frekuensi')
                    .sort_values('frekuensi', ascending=False).head(10))

print(f"Kombinasi paling umum       : K={mode_combo[0]}, V={mode_combo[1]}")
print("\nTOP 10 Kombinasi Konsonan–Vokal Terbanyak:")
print("-" * 90)
for _, row in top_combinations.iterrows():
    bar = '█' * (int(row['frekuensi']) // 5)
    print(f"K:{int(row['konsonan'])} V:{int(row['vokal'])} │ {bar} ({int(row['frekuensi'])} kata)")

# Visualisasi Heatmap
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle('Heatmap Analisis Konsonan–Vokal', fontsize=16, fontweight='bold', y=1.02)

sns.heatmap(pivot_frequency, annot=True, fmt='d', cmap='YlOrRd', ax=axes[0],
            linewidths=0.5, cbar_kws={'label': 'Frekuensi'})
axes[0].set_title('Frekuensi Total Kombinasi', fontsize=14, fontweight='bold', pad=15)
axes[0].set_xlabel('Jumlah Konsonan', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Jumlah Vokal', fontsize=12, fontweight='bold')

sns.heatmap(pivot_count, annot=True, fmt='d', cmap='Blues', ax=axes[1],
            linewidths=0.5, cbar_kws={'label': 'Cacah Kata Unik'})
axes[1].set_title('Cacah Kata Unik per Kombinasi', fontsize=14, fontweight='bold', pad=15)
axes[1].set_xlabel('Jumlah Konsonan', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Jumlah Vokal', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

# Interpretasi & Insight 
print("\n INTERPRETASI & INSIGHT")
print("-" * 90)
print(f"""
1) Pola Dominan: Kombinasi K={mode_combo[0]}, V={mode_combo[1]} paling sering muncul. 
   Ini mencerminkan gaya bahasa Indonesia yang cenderung bersuku kata terbuka (konsonan-vokal).

2) Perbedaan Dua Tabel:
   - Frekuensi Total menghitung semua kemunculan kata (termasuk pengulangan istilah kunci).
   - Cacah Unik menghitung setiap kata berbeda hanya 1×. 
   Selisih yang tampak menandakan adanya istilah yang sering diulang pada teks akademik.

3) Rasio C:V ≈ {ratio_cv:.2f}:1 menunjukkan keseimbangan fonetik yang wajar; 
   tidak ada dominasi salah satu yang berlebihan — baik untuk keterbacaan ilmiah.

4) Implikasi:
   Dominasi kombinasi konsonan menengah dan vokal 1–2 mengindikasikan 
   banyak kata bermorfem ringkas namun informatif (misal istilah teknis).
""")
print()

# =====================================
# BAGIAN 3: JUMLAH KATA DALAM KALIMAT 
# =====================================
print("╔" + "="*88 + "╗")
print("║" + " "*23 + "3. ANALISIS JUMLAH KATA DALAM KALIMAT" + " "*28 + "║")
print("╚" + "="*88 + "╝")
print()

sentences = extract_sentences(full_text)
words_per_sentence = [len(extract_words(s)) for s in sentences]

# Statistik
mean_wps = float(np.mean(words_per_sentence)) if sentences else 0.0
median_wps = float(np.median(words_per_sentence)) if sentences else 0.0
std_wps = float(np.std(words_per_sentence)) if sentences else 0.0
min_wps = min(words_per_sentence) if sentences else 0
max_wps = max(words_per_sentence) if sentences else 0

print("STATISTIK PANJANG KALIMAT:")
print("-" * 90)
print(f"Total kalimat dianalisis    : {len(sentences)}")
print(f"Rata-rata kata/kalimat      : {mean_wps:.2f} kata")
print(f"Median kata/kalimat         : {median_wps:.0f} kata")
print(f"Standar deviasi             : {std_wps:.2f}")
print(f"Kalimat terpendek           : {min_wps} kata")
print(f"Kalimat terpanjang          : {max_wps} kata")
print(f"Range                       : {max_wps - min_wps} kata")

# Histogram jumlah kata per kalimat
plt.figure(figsize=(9, 5))
bins_wps = range(1, (max_wps or 1) + 2)
plt.hist(words_per_sentence, bins=bins_wps, edgecolor='black', alpha=0.7, color='lightcoral')

plt.title("Histogram: Jumlah Kata per Kalimat", fontsize=13, fontweight='bold')
plt.xlabel("Jumlah Kata", fontsize=11)
plt.ylabel("Frekuensi Kalimat", fontsize=11)

# 🔹 Tambahan agar angka di bawah tidak dempet tapi tetap informatif
step = 5 if max_wps > 20 else 1  # tampilkan setiap 5 angka kalau data panjang
plt.xticks(range(0, max_wps + step, step), fontsize=10)

# 🔹 Sedikit jarak antar batang agar tidak rapat
plt.xlim(0, max_wps + 2)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()


# Interpretasi & Insight 
print("\n INTERPRETASI & INSIGHT")
print("-" * 90)
print(f"""
1) Kompleksitas Kalimat: Rata-rata {mean_wps:.1f} kata per kalimat menandakan 
   gaya akademik yang informatif namun masih terbaca. Nilai median {median_wps:.0f} 
   menunjukkan sebaran tidak terlalu timpang.

2) Variabilitas: Standar deviasi {std_wps:.2f} menunjukkan variasi panjang kalimat 
   yang baik. Ada kalimat singkat untuk penekanan dan ada kalimat panjang untuk elaborasi.

3) Implikasi Keterbacaan: Jumlah kata ada di rentang {min_wps}–{max_wps} kata per kalimat yang dimana konsisten dengan 
   rekomendasi penulisan ilmiah (umumnya 15–25 kata/kalimat untuk kelancaran pemahaman).
""")
print()

# =========================================
# BAGIAN 4: JUMLAH KALIMAT DALAM PARAGRAF 
# =========================================
print("╔" + "="*88 + "╗")
print("║" + " "*21 + "4. ANALISIS JUMLAH KALIMAT DALAM PARAGRAF" + " "*26 + "║")
print("╚" + "="*88 + "╝")
print()

sentences_per_paragraph = [len(extract_sentences(p)) for p in all_paragraphs]

# Statistik
mean_spp = float(np.mean(sentences_per_paragraph)) if sentences_per_paragraph else 0.0
median_spp = float(np.median(sentences_per_paragraph)) if sentences_per_paragraph else 0.0
std_spp = float(np.std(sentences_per_paragraph)) if sentences_per_paragraph else 0.0
min_spp = min(sentences_per_paragraph) if sentences_per_paragraph else 0
max_spp = max(sentences_per_paragraph) if sentences_per_paragraph else 0

print("STATISTIK PARAGRAF:")
print("-" * 90)
print(f"Total paragraf dianalisis   : {len(all_paragraphs)}")
print(f"Rata-rata kalimat/paragraf  : {mean_spp:.2f} kalimat")
print(f"Median kalimat/paragraf     : {median_spp:.0f} kalimat")
print(f"Standar deviasi             : {std_spp:.2f}")
print(f"Paragraf terpendek          : {min_spp} kalimat")
print(f"Paragraf terpanjang         : {max_spp} kalimat")

# Histogram jumlah kalimat per paragraf 
plt.figure(figsize=(9,4.5))
bins_spp = range(1, (max_spp or 1) + 2)
plt.hist(sentences_per_paragraph, bins=bins_spp, edgecolor='black', alpha=0.8)
plt.title("Histogram: Jumlah Kalimat per Paragraf")
plt.xlabel("Jumlah Kalimat dalam Paragraf")
plt.ylabel("Frekuensi Paragraf")
plt.xticks(bins_spp)
plt.tight_layout()
plt.show()

# Interpretasi & Insight 
labels = ['Abstrak', 'Pendahuluan-p1', 'Pendahuluan-p2', 'Kesimpulan-p1', 'Kesimpulan-p2']
print("\n INTERPRETASI & INSIGHT ")
print("-" * 90)
print(f"""
1) Struktur Paragraf: Rata-rata {mean_spp:.1f} kalimat/paragraf menunjukkan fokus ide 
   per paragraf terjaga. Median {median_spp:.0f} menandakan setengah paragraf berada 
   di sekitar nilai tersebut (tidak ekstrim).

2) Persebaran: Rentang {min_spp}–{max_spp} kalimat memperlihatkan variasi fungsi:
   - Paragraf pembuka (abstrak/pendahuluan) cenderung lebih panjang untuk konteks.
   - Paragraf kesimpulan lebih ringkas dan langsung ke poin.

3) Implikasi Komposisi: Histogram yang terkonsentrasi pada 2–4 kalimat/paragraf 
   ideal untuk menyeimbangkan kedalaman argumen dan keterbacaan.
""")
print()

# =============================
# BAGIAN 5: ANALISIS TAMBAHAN 
# =============================

print("╔" + "="*66 + "╗")
print("║" + " "*21 + "4. ANALISIS TAMBAHAN" + " "*25 + "║")
print("╚" + "="*66 + "╝")
print()

# Analisis 1: Distribusi Panjang Kata (Kata Unik)
word_lengths = [len(w) for w in df_unique['word']]

plt.figure(figsize=(9,5))
plt.hist(word_lengths, bins=range(1, max(word_lengths)+2), edgecolor='black', alpha=0.75, color='lightcoral')
plt.title("Distribusi Panjang Kata (Kata Unik)", fontsize=13, fontweight='bold')
plt.xlabel("Panjang Kata (jumlah huruf)")
plt.ylabel("Frekuensi Kata")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("INSIGHT & INTERPRETASI")
print("-"*90)
print(f"""
Rata-rata panjang kata: {np.mean(word_lengths):.2f} huruf
Panjang kata paling sering muncul: {Counter(word_lengths).most_common(1)[0][0]} huruf

Interpretasi:
Sebagian besar kata berada di kisaran 5–7 huruf, menunjukkan karakteristik umum bahasa Indonesia.
Jika banyak kata panjang (>10 huruf), berarti artikel menggunakan istilah teknis seperti 'penerapan' atau 'algoritma'.
Distribusi yang seimbang menandakan teks mudah dibaca namun tetap akademik.
""")

# Analisis 2: Frekuensi Kata (Top 15)
top_words = Counter(df_words['word']).most_common(15)
top_df = pd.DataFrame(top_words, columns=['Kata', 'Frekuensi'])

plt.figure(figsize=(9,5))
sns.barplot(data=top_df, x='Frekuensi', y='Kata', color = 'lightcoral')
plt.title("15 Kata Paling Sering Muncul dalam Artikel", fontsize=13, fontweight='bold')
plt.xlabel("Frekuensi Kemunculan")
plt.ylabel("Kata")
plt.tight_layout()
plt.show()

print("INSIGHT & INTERPRETASI")
print("-"*90)
print("""
Interpretasi:
Kata seperti “produk”, “stok”, “algoritma”, “data”, dan “herbal” sering muncul.
Hal ini menegaskan fokus utama artikel pada penerapan algoritma C4.5 untuk memprediksi stok produk herbal HNI-HPAI.
Frekuensi kata kunci memperlihatkan bahwa penelitian bersifat aplikatif, menekankan implementasi nyata.
""")

# Analisis 3: Rata-rata Panjang Kalimat per Paragraf 
avg_sentence_len_per_para = []
for para in all_paragraphs:
    sents = extract_sentences(para)
    avg_len = np.mean([len(extract_words(s)) for s in sents]) if sents else 0
    avg_sentence_len_per_para.append(avg_len)

plt.figure(figsize=(9,4.5))
plt.bar(range(1, len(all_paragraphs)+1), avg_sentence_len_per_para, color='lightcoral', edgecolor='black')
plt.title("Rata-rata Panjang Kalimat per Paragraf", fontsize=13, fontweight='bold')
plt.xlabel("Paragraf ke-")
plt.ylabel("Rata-rata Jumlah Kata per Kalimat")
plt.xticks(range(1, len(all_paragraphs)+1), ['Abs', 'Pend-1', 'Pend-2', 'Kes-1', 'Kes-2'])
plt.tight_layout()
plt.show()

print("INSIGHT & INTERPRETASI")
print("-"*90)
for i, val in enumerate(avg_sentence_len_per_para, 1):
    print(f"Paragraf {i}: rata-rata {val:.2f} kata per kalimat")

print("""
Interpretasi:
Abstrak dan Pendahuluan memiliki kalimat lebih panjang karena menjelaskan konteks dan teori.
Kesimpulan lebih singkat dan langsung pada hasil, menandakan struktur ilmiah yang efektif.
Perbedaan ini menunjukkan keseimbangan antara bagian penjelasan dan ringkasan hasil.
""")

# ===========================
# SINTESIS AKHIR (Semua Bagian)
# ===========================
print("KESIMPULAN")
print(f"""
1) Struktur Huruf dan Pola Bahasa (Bagian 2):
   - Kombinasi konsonan–vokal paling dominan: K={mode_combo[0]}, V={mode_combo[1]}.
     Pola ini khas bahasa Indonesia yang bersuku kata terbuka, sehingga memudahkan pelafalan.
   - Selisih antara frekuensi total dan cacah unik menunjukkan adanya istilah inti yang diulang,
     seperti 'produk', 'stok', dan 'algoritma', yang menjadi fokus utama penelitian.

2) Pola Kalimat (Bagian 3):
   - Rata-rata {mean_wps:.1f} kata per kalimat dan median {median_wps:.0f} menandakan
     struktur kalimat ilmiah yang padat informasi, namun tetap komunikatif.
   - Rentang {min_wps}–{max_wps} kata/kalimat serta standar deviasi {std_wps:.2f} menunjukkan
     variasi panjang kalimat yang baik: ada kalimat panjang untuk argumentasi dan
     kalimat singkat untuk penegasan ide.

3) Komposisi Paragraf (Bagian 4):
   - Rata-rata {mean_spp:.1f} kalimat per paragraf menunjukkan tiap paragraf berfokus pada satu ide pokok.
   - Paragraf pendahuluan cenderung lebih panjang karena berisi konteks dan teori,
     sementara paragraf kesimpulan lebih ringkas dan langsung pada hasil.
   - Variasi panjang ini mencerminkan struktur penulisan ilmiah yang efektif.

4) Analisis Tambahan (Bagian 5):
   - Distribusi panjang kata menunjukkan mayoritas berada di kisaran 5–7 huruf,
     menandakan kombinasi kata umum dan istilah teknis yang seimbang.
   - Frekuensi kata menguatkan bahwa topik utama penelitian berpusat pada
     penerapan algoritma C4.5 dalam konteks stok produk herbal HNI-HPAI.
   - Rata-rata panjang kalimat per paragraf mengonfirmasi bahwa bagian pendahuluan
     lebih elaboratif, sedangkan kesimpulan lebih padat dan efisien.
     
Kesimpulan Umum:
   Secara keseluruhan, artikel ini memiliki struktur penulisan yang baik. Kalimatnya informatif,
   paragraf terfokus, dan distribusi kata yang seimbang. Dari analisis kuantitatif ini dapat
   disimpulkan bahwa penulis konsisten menggunakan gaya ilmiah yang sistematis, jelas, dan efektif.
""")

print("\n" + "="*90)
print(" "*35 + "ANALISIS SELESAI")
print("="*90)
```
